```python
class Number(object):
    def reverse(self, n):
        if n // 10:
            return 10 ** (len(str((n))) - 1) * (n % 10) + self.reverse(n // 10)
        else:
            return n
        
print('Created')
```

```python
n = Number()
print(n)
rev_n = n.reverse(1234)
print(rev_n)
print(rev_n + 9)
```
